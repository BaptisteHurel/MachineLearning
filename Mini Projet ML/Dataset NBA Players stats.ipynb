### 1.3) CSV / Affichage de seasons_stats

```python
seasons_stats = pd.read_csv("seasons_stats.csv")
```

```python
#seasons_stats.drop(['Unnamed: 0'], axis='columns', inplace = True)
seasons_stats.head()
```

```python
seasons_stats2 = seasons_stats[['Player','Pos','G','ORB','DRB','AST','PTS']]
```

```python
seasons_stats2.fillna(0, inplace=True)
```

```python
seasons_stats2
```

```python
duplicas=seasons_stats2.duplicated(keep=False,subset='Player')
doublons=seasons_stats2[duplicas]
```

```python
grouped=doublons.groupby(['Player'])
grouped.head()
```

```python
seasons_stats2.rename(columns={'G': 'Game'}, inplace=True)
seasons_stats2.rename(columns={'ORB': 'Offensive Rebond'}, inplace=True)
seasons_stats2.rename(columns={'DRB': 'Defensive Rebond'}, inplace=True)
```

```python
seasons_stats2
```

### 2) Partie régression linéaire

```python
X=seasons_stats2[['PTS','Offensive Rebond','Defensive Rebond','AST']]
Y=seasons_stats2['Game']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=40)
```

```python
X_train
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lin_model = LinearRegression()
```

```python
lin_model.fit(X_train, Y_train)
```

```python
lin_model.predict(X_test)
```

```python
Y_test
```

```python
lin_model.score(X_test, Y_test)
```
